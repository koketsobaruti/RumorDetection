### The differences in Friend_count distribution between rumors and non-rumors

```python
import numpy as np
import matplotlib.pyplot as plt
# Extract friends_count for both rumor types
true_rumors = r_df['friends_count']
false_rumors = nr_df['friends_count']
# Generate example data (uniformly distributed)

# Step 1: Sort the data
tr_sorted_data = np.sort(true_rumors)
fr_sorted_data = np.sort(false_rumors)
# Step 2: Calculate the CDF
cdf_tr = np.linspace(0, 1, len(tr_sorted_data))
cdf_fr = np.linspace(0, 1, len(fr_sorted_data))

# Step 3: Plot the CDF with logarithmic x-axis
plt.figure(figsize=(8, 6))
plt.plot(tr_sorted_data, cdf_tr, label='CDF True Rumor')
plt.plot(fr_sorted_data, cdf_fr, label='CDF False Rumor')
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.xlabel('Value (Log Scale)')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the CCDF for friends_count
def ccdf(data):
    sorted_data = np.sort(data)
    ccdf_values = 1.0 - np.arange(len(sorted_data)) / len(sorted_data)
    return sorted_data, ccdf_values

# Extract friends_count for both rumor types
true_rumors = r_df['friends_count']
false_rumors = nr_df['friends_count']
# Normalize friends_count using MinMaxScaler
scaler = MinMaxScaler()

tr = np.array(true_rumors).reshape(-1,1)
fr = np.array(false_rumors).reshape(-1,1)
normalized_tr = scaler.fit_transform(tr).flatten()
normalized_fr = scaler.fit_transform(fr).flatten()
# Calculate CCDF for both categories
true_rumors_sorted, true_rumors_ccdf = ccdf(normalized_tr)
false_rumors_sorted, false_rumors_ccdf = ccdf(normalized_fr)

# Plotting the CCDF
plt.figure(figsize=(10, 6))
plt.plot(true_rumors_sorted, true_rumors_ccdf, label='True Rumors')
plt.plot(false_rumors_sorted, false_rumors_ccdf, label='False Rumors')
plt.xscale('log')
plt.xlabel('Friends Count')
plt.ylabel('Complementary Cumulative Distribution Function (CCDF)')
plt.title('CCDF of Friends Count for True vs False Rumors')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

# Identify exact duplicates
exact_duplicates = rumor_data[rumor_data.duplicated(subset=['tweet_text'], keep=False)]

# Identify similar strings
def find_similar_rumors(df, threshold=80):
    similar_pairs = []
    for i, rumor in df.iterrows():
        # Compare each rumor to all other rumors
        matches = process.extract(rumor['tweet_text'], df['tweet_text'], scorer=fuzz.ratio)
        for match in matches:
            if match[1] >= threshold and rumor['tweet_text'] != match[0]:
                similar_pairs.append((rumor['tweet_text'], match[0], match[1]))
    return similar_pairs

similar_rumors = find_similar_rumors(rumor_data)

# Display results
print("Exact Duplicates:")
display(exact_duplicates)

#print("\nSimilar Rumors (with similarity score):")
#for pair in similar_rumors:
    #print(f"{pair[0]} <-> {pair[1]} : {pair[2]}")

# Visualizing counts
count_exact_duplicates = len(exact_duplicates)
count_similar_rumors = len(similar_rumors)

categories = ['Exact Duplicates', 'Similar Rumors']
counts = [count_exact_duplicates, count_similar_rumors]

fig = go.Figure()

fig.add_trace(go.Bar(
    x=categories,
    y=counts,
    text=counts,
    textposition='auto',
    marker=dict(color=['lightskyblue', 'lightcoral']),
    width=0.5
))

fig.update_layout(
    title='Count of Exact Duplicates and Similar Rumors',
    xaxis_title='Category',
    yaxis_title='Count',
    height=400,
    width=700
)

fig.show()
```

### Difference of Emotions Expressed In Rumors and Non-Rumors

```python
def aggregate_emotions(emotion_scores):
    positive_emotions = ['anticipation', 'trust', 'surprise', 'positive', 'joy']
    negative_emotions = ['fear', 'anger', 'negative', 'sadness', 'disgust']

    positive_sum = sum(emotion_scores[emotion] for emotion in positive_emotions if emotion in emotion_scores)
    negative_sum = sum(emotion_scores[emotion] for emotion in negative_emotions if emotion in emotion_scores)

    # Method 1: Difference between positive and negative sums
    overall_emotion_score_difference = positive_sum - negative_sum

    # Method 2: Weighted sum (assuming equal weight here for simplicity)
    overall_emotion_score_weighted = (positive_sum - negative_sum) / (positive_sum + negative_sum + 1e-10)

    # Method 3: Positive to Negative ratio
    positive_negative_ratio = (positive_sum + 1e-10) / (negative_sum + 1e-10)

    return {
        'difference': overall_emotion_score_difference,
        'weighted': overall_emotion_score_weighted,
        'ratio': positive_negative_ratio
    }
```

```python
from nrclex import NRCLex
rumor_tweets = rumor_data['tweet_text']
rumor_dict = {}
for i in rumor_tweets.index:
    tweet = rumor_tweets[i]
    obj_emotions = NRCLex(tweet).affect_frequencies
    aggr_emotions = aggregate_emotions(obj_emotions)
    rumor_dict.update({i:aggr_emotions})

print(rumor_dict[0])
```

```python
def process_in_batches(texts, batch_size=32):
    elmo_embed_np = []
    total_batches = (len(texts) + batch_size - 1) // batch_size  # Calculate total number of batches
    with tqdm(total=total_batches, desc='Processing Batches') as pbar:
        for i in range(0, len(texts), batch_size):
            batch_texts = texts[i:i + batch_size]
            batch_embeddings = elmo_embeddings(batch_texts)
            # Average the embeddings for each text
            batch_embeddings_np = [np.mean(embed.numpy(), axis=0) for embed in batch_embeddings]
            for embed in batch_embeddings_np:
                if embed.shape != (1024,):
                    print(f"Unexpected shape found: {embed.shape}")
            elmo_embed_np.extend(batch_embeddings_np)
            pbar.update(1)  # Update progress bar
    return np.array(elmo_embed_np)
```

```python
# Function to convert text to ELMo embeddings
def elmo_embeddings(text_list):
    embeddings = elmo.signatures['default'](tf.constant(text_list))['elmo']
    return embeddings
```
